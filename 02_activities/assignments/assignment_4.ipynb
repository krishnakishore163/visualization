Visualization 2: Bar chart of Washroom types and Statuses

```python
# Count the number of each type of washroom and its status
type_status_counts = data.groupby(['type', 'Status']).size().unstack().fillna(0)

# Plot the bar chart
type_status_counts.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Distribution of Washroom Types and Their Statuses')
plt.xlabel('Washroom Type')
plt.ylabel('Count')
plt.legend(title='Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
